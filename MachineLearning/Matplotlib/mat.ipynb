```python
x1=np.array([1,2,3,4])
y1=np.array([4,7,8,10])
plt.plot(x1,y1,"o--b")
```

```python
plt.plot(x1,y1,"o--r")
```

```python
plt.plot(x1,y1,"s:r")
```

```python
plt.plot(x1,y1,"o--y")
```

```python
plt.plot(x1,y1,marker='o',ms=15)
#ms marker size
```

```python
plt.plot(x1,y1,marker='o',ms=15,mfc='r')
#ms marker size,mfc>marker face color
```

```python
plt.plot(x1,y1,marker='o',ms=15,mec='g',mfc='r')
#ms marker size
```

```python
plt.plot(x1,y1,marker='o',linestyle='dotted')
#ms marker size
```

```python
plt.plot(x1,y1,marker='o',linestyle='dashed',linewidth=4)
#ms marker size
```
