```python
import matplotlib.pyplot as plt
height=[150,155,156,157]
weight=[45,50,51,58]
plt.plot(height,weight)
plt.xlabel('height in cm',color='r',fontsize=10)
plt.ylabel('weight in kg')
plt.title("Height-Weight Graph")
```

```python
velocity=[150,155,156,157]
displacement=[45,50,51,58]
plt.plot(velocity,displacement)
plt.xlabel('velocity in m/s',color='r',fontsize=10)
plt.ylabel('displacement in m',color='g',fontsize=10)
plt.title("velocity-displacement Graph")
plt.grid()#can define axix if needed plt.grid(axis='x/y')
```

```python
#subplot
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title('Graph1')
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('Graph2')
plt.suptitle('Graph1-Graph2')
```
